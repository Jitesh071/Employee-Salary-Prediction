Label Encoding:

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data['Gender']=encoder.fit_transform(data['Gender'])
data['Education Level']=encoder.fit_transform(data['Education Level'])
data['Job Title']=encoder.fit_transform(data['Job Title'])
data.head(10)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
data['Age']=scaler.fit_transform(data[['Age']])
data['Years of Experience']=scaler.fit_transform(data[['Years of Experience']])
display(data.head())
```

```python
x=data.drop(columns=['Salary'])
y=data['Salary']
x
```

Traning and Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.2,random_state=43)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape,y_test.shape
```

Model Development

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

Modal Predictions:

```python
prd=lr.predict(x_test)
prd
```

```python
df = pd.DataFrame({'y_Actual' :y_test, 'y_Predicted' : prd})
df['Error']=df['y_Actual'] - df['y_Predicted']
df['abs_error'] = abs(df['Error'])
df
```

```python
Mean_absolute_error = df['abs_error'].mean()
Mean_absolute_error
```

Modal Evaluation:

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

Modal Accuracy:

```python
r2_score(y_test,prd)*100
```

```python
print(f'Accuracy of the modal = {round(r2_score(y_test,prd),4)*100}%')
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the trained model
model = joblib.load("lr.joblib")

# Initialize encoders and scaler with expected values from training data
# These lists should contain all possible values for each categorical feature
gender_encoder = LabelEncoder()
gender_encoder.fit(["Female", "Male"])

education_encoder = LabelEncoder()
education_encoder.fit(["Bachelor's", "Master's", "PhD"])

job_title_encoder = LabelEncoder()
job_title_encoder.fit(["Software Engineer", "Senior Manager", "Director", "Data Analyst", "Sales Associate"])

# Initialize and fit the scaler on representative data (e.g., dummy data with min/max ranges or actual training data if available)
# For simplicity, using dummy data based on observed ranges in the original notebook
scaler = StandardScaler()
# Fit on dummy data that covers the range of values seen during training
dummy_data = pd.DataFrame({
    'Age': [18, 65],
    'Years of Experience': [0, 40]
})
scaler.fit(dummy_data)


st.set_page_config(page_title="Employee Salary Prediction", page_icon="💼", layout="centered")

st.title("💼 Employee Salary Prediction App")
st.markdown("Predict employee salary based on input features.")

# Sidebar inputs (these must match your training feature columns)
st.sidebar.header("Input Employee Details")

Age = st.sidebar.slider("Age", 18, 65, 30)
education_level = st.sidebar.selectbox("Education Level", [
    "Bachelor's", "Master's", "PhD"
])
Gender = st.sidebar.selectbox("Gender", [
    "Male","Female"
])
Job_Title = st.sidebar.selectbox("Job Title", [
    "Software Engineer", "Senior Manager", "Director", "Data Analyst","Sales Associate"
])
years_of_experience = st.sidebar.slider("Years of Experience", 0, 40, 5)

# Build input DataFrame for single prediction
input_df = pd.DataFrame({
    'Age': [Age],
    'Gender': [Gender],
    'Education Level': [education_level],
    'Job Title': [Job_Title],
    'Years of Experience': [years_of_experience]
})


st.write("### 🔎 Input Data")
st.write(input_df)

# Predict button
if st.button("Predict Salary"):
    # Preprocess input data
    input_df['Gender'] = gender_encoder.transform(input_df['Gender'])
    input_df['Education Level'] = education_encoder.transform(input_df['Education Level'])
    input_df['Job Title'] = job_title_encoder.transform(input_df['Job Title'])

    # Scale numerical features
    input_df[['Age', 'Years of Experience']] = scaler.transform(input_df[['Age', 'Years of Experience']])

    # Ensure column order matches training data
    # Based on the training code, the order was 'Age', 'Gender', 'Education Level', 'Job Title', 'Years of Experience'
    input_df = input_df[['Age', 'Gender', 'Education Level', 'Job Title', 'Years of Experience']]


    prediction = model.predict(input_df)
    st.success(f"✅ Predicted Salary: ${prediction[0]:,.2f}")

# Batch prediction
st.markdown("---")
st.markdown("#### 📂 Batch Prediction")
uploaded_file = st.file_uploader("Upload a CSV file for batch prediction", type="csv")

if uploaded_file is not None:
    batch_data = pd.read_csv(uploaded_file)
    st.write("Uploaded data preview:", batch_data.head())

    # Preprocess batch data
    batch_data['Gender'] = gender_encoder.transform(batch_data['Gender'])
    batch_data['Education Level'] = education_encoder.transform(batch_data['Education Level'])
    batch_data['Job Title'] = job_title_encoder.transform(batch_data['Job Title'])

    batch_data[['Age', 'Years of Experience']] = scaler.transform(batch_data[['Age', 'Years of Experience']])

    # Ensure column order matches training data
    batch_data = batch_data[['Age', 'Gender', 'Education Level', 'Job Title', 'Years of Experience']]

    batch_preds = model.predict(batch_data)
    batch_data['Predicted Salary'] = batch_preds
    st.write("✅ Predictions:")
    st.write(batch_data.head())
    csv = batch_data.to_csv(index=False).encode('utf-8')
    st.download_button("Download Predictions CSV", csv, file_name='predicted_salaries.csv', mime='text/csv')
```
